```python
import matplotlib.pyplot as plt
from flyplotlib import add_fly

fig, ax = plt.subplots()
add_fly(xy=(0, 0), ax=ax)
add_fly(xy=(2, -0.5), rotation=30, length=2, alpha=0.3, ax=ax)
add_fly(xy=(3.5, 0), rotation=-45, grayscale=True, ax=ax)
add_fly(xy=(6.5, 0), ax=ax, alpha=0.5, exclude=["leg", "stripe", "haltere"])
ax.set_aspect(1)
```

```python
import numpy as np
from matplotlib.patches import Wedge

fig, (ax0, ax1) = plt.subplots(2, 1, figsize=(9, 6))
add_fly(xy=(-3, 0), rotation=45, ax=ax0)
add_fly(xy=(3, 0), ax=ax0)
ax0.plot([-2, -3, 3], [1, 0, 0], "k")
ax0.add_artist(Wedge((-3, 0), 1, 0, 45, alpha=0.5))

for x in np.arange(np.pi * 4, 0, -np.pi / 8):
    rotation = np.rad2deg(np.arctan(np.cos(x)))
    add_fly(xy=(x, np.sin(x)), rotation=rotation)

for ax in (ax0, ax1):
    ax.set_aspect(1)

fig.tight_layout()
```
